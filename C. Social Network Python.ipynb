```python
G_SNA = nx.Graph()
for index, row in df.iterrows():
    #print(index, row["EmpID"],row["SupID"] )
    G_SNA.add_edge(row["EmpID"], row["SupID"], weight=1)
```

```python
print(nx.info(G_SNA))
```

```python
G_SNA = nx.Graph() 
for index, row in df.iterrows():
    G_SNA.add_edge(row["EmpID"], row["SupID"])

#plt.plot() 
plt.figure(figsize=(10,10))

nx.draw(G_SNA, pos = nx.circular_layout(G_SNA), with_labels=True, node_size= 2000,
        node_color = 'r', 
        edge_color = 'b')
```

```python
G_SNA = nx.Graph() 
for index, row in df.iterrows():
    G_SNA.add_edge(row["EmpID"], row["SupID"])

#plt.plot() 
plt.figure(figsize=(18,18))

nx.draw(G_SNA, pos = nx.spring_layout(G_SNA), with_labels=True, node_size= 2000,
        node_color = 'r', 
        edge_color = 'b')
```
