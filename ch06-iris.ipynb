```r
install.packages('rpart')
install.packages("rpart.plot")


library(rpart)
library(rpart.plot)
```

```r
model.ID3 <- rpart( Species~., 
                    data=iris, 
                    method="class", 
                    parms=list(split="information"))
model.ID3
```

```r
rpart.plot( model.ID3, branch=1, type=1, fallen.leaves=T, cex=1, sub="決策樹模型-ID3")
```

```r
library(rpart)
model.cart<-rpart(Species ~., 
                  data=iris, 
                  method="class", 
                  parms=list( split="gini" ) ) # 使用CART算法的時候, split = “gini”
model.cart
```

```r
rpart.plot( model.cart, branch=1, type=2, fallen.leaves=F, cex=1, sub="決策樹模型-CART")
```
