```r
data = data.frame(x=1:20,
                  y=c(3,4,8,2,6,10,12,13,15,14,17,18,20,17,21,22,25,30,29,31))
```

```r
library(e1071)
```

```r
plot(data$x, data$y, pch=16, xlab="X", ylab="Y")
```

# 傳統線性回歸

```r
model <- lm(y ~ x , data) 
```

```r
lm.pred = predict(model, data)
```

```r
plot(data$x, data$y, pch=16, xlab="X", ylab="Y")
points(lm.pred, pch=2, col="red")
abline(model, col="red")
```

# 用SVM

```r
model1 <- svm(y ~ x , data)
```

```r
svr.pred = predict(model1, data)
```

```r
plot(data$x, data$y, pch=16, xlab="X", ylab="Y")
points(svr.pred, pch=4, col="blue")
```

# 比較線性迴歸與SVM的差異

```r
plot(data$x, data$y, pch=16, xlab="X", ylab="Y")
points(lm.pred, pch=2, col="red")
points(svr.pred, pch=4, col="blue")
```

```r
c(sqrt(mean((data$y - lm.pred)^2)),  sqrt(mean((data$y - svr.pred)^2)) )
```

```r
set.seed(50)

train=sample(1:150,105)		#訓練集 7成
train
```

```r
svm_model = svm(Species ~ ., data=iris[train,])  #訓練集來訓練模型
```

```r
svm_model
```

```r
pred <- predict(svm_model, iris[-train,])		#預測 測試集的花瓣寬
```

```r
cm <- table(pred, iris[-train,]$Species)
```

```r
cm
```

```r
SVMaccuracy <- sum(diag(cm)) / sum(cm)
```
