```r
set.seed(50)
```

```r
n <- nrow(iris)
n
```

```r
t_idx <- sample(seq_len(n), size = round(0.7 * n))
```

```r
t_idx
```

```r
traindata <- iris[t_idx,]
```

```r
traindata
```

```r
testdata <- iris[ - t_idx,]
```

```r
testdata
```

```r
install.packages('e1071')
```

```r
library(e1071)
```

```r
install.packages("rminer") 
library(rminer)
```

```r
nbModel <- naiveBayes(Species ~ ., data = traindata)   #     
```

```r
nbModel
```

```r
results <- predict(nbModel, testdata)  #     
```

```r
results
```

```r
cm <- table(x = testdata$Species, y = results)
```

```r
cm
```

```r
naiveBayesaccuracy <- sum(diag(cm)) / sum(cm)
```

```r
naiveBayesaccuracy
```
