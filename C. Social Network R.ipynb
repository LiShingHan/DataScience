```r
#install.packages("igraph")
install.packages(c("igraph"))
library(igraph)
```

```r
purhead<- read.csv("https://raw.githubusercontent.com/LiShingHan/DataScience/main/DataSet/purhead.csv", fileEncoding = "big-5")
```

```r
head(purhead)
```

```r
newframe<-data.frame(purhead$EmpID,purhead$SupID)
```

```r
netgraph<-graph.data.frame(newframe, directed = FALSE)
netgraph
```

```r
plot(netgraph, layout=layout.fruchterman.reingold)
```

```r
plot(netgraph,layout=layout.circle)
```
